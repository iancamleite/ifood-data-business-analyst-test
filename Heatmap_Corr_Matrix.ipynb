```python
# Load the Automobile dataset.
# This gets a cleaned version of UCI dataset found at http://archive.ics.uci.edu/ml/datasets/automobile

data = pd.read_csv('https://raw.githubusercontent.com/iancamleite/ifood-data-business-analyst-test/master/ml_project1_data.csv')
```
